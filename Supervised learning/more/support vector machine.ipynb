```python
import pandas as pd 
from sklearn.datasets import load_iris 
iris = load_iris()
```

```python
dir(iris)
```

```python
iris.feature_names
```

```python
df = pd.DataFrame(iris.data ,columns = iris.feature_names)
df.head()
```

```python
df['target'] = iris.target
df
```

```python
df[df.target==0]
```

```python
import matplotlib.pyplot as plt
```

```python
df0 = df[df.target==0]
df1 = df[df.target==1]
df2 = df[df.target==2]
#to assign a seperate data frame
```

```python
plt.scatter(df0['sepal length (cm)'],df0['sepal width (cm)'],c='red',marker = '+')
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'],c='green',marker = '.')



```

```python
plt.scatter(df0['petal length (cm)'],df0['petal width (cm)'],c='red',marker = '+')
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'],c='green',marker = '.')


```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop('target',axis = 'columns')
y = df.target
```

```python
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size = 0.2)
```

```python
from sklearn.svm import SVC
```

```python
model = SVC()
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
